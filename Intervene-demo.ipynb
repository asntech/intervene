## Intervene interactive mode

```python
#load intervene module
import intervene as iv
```

**Prepare input data**

A list of files (genomic regions/name lists)

```python
file_list = ['./intervene/example_data/dbSUPER_mm9/Liver.bed',
             './intervene/example_data/dbSUPER_mm9/Cortex.bed',
             './intervene/example_data/dbSUPER_mm9/Spleen.bed']
```

```python
#get weighted classical Venn
weights, label_names, output_file = iv.venn(
    input=file_list,
    type='genomic', 
    style='Classical',
    faces=True)
#check the output
from IPython.display import Image
Image(filename=output_file)
```

```python
#get not weighted Venn by setting weighted=False
weights, label_names, output_file = iv.venn(input=file_list, type='genomic', style='Classical', weighted=False)
Image(filename=output_file)
```

```python
#get a weighted 4-way ChowRuskey Venn by setting style=ChowRuskey
file_list2 = ['./intervene/example_data/dbSUPER_mm9/Liver.bed',
             './intervene/example_data/dbSUPER_mm9/Cortex.bed',
             './intervene/example_data/dbSUPER_mm9/Spleen.bed',
             './intervene/example_data/dbSUPER_mm9/Kidney.bed']
weights, label_names, output_file = iv.venn(input=file_list2, type='genomic', style='ChowRuskey')
Image(filename=output_file)
```

```python
weights, label_names, png_file = iv.upset(file_list, type='genomic')
```
